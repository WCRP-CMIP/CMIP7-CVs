```python
import pandas as pd 
import json, os, glob, re

df = pd.read_csv('dr_experiments.csv')
df = df[df['Count (Opportunity)'] > 0]
df['Experiment Group'] = [[re.split('[-_]', x)[0].replace('fast','fast-track').lower() for x in y.split(',') if 'all-' not in x] for y in df['Experiment Group']]
```

```python
for _,r in df.iterrows():
    experiment = r[' Experiment']
    file = f'experiment/{experiment.lower()}.json'
    
    if not os.path.exists(file):
        print('aaa',experiment)
        
        template = {
                "id": f"{experiment.lower()}",
                "validation-key": f"{experiment}",
                "ui-label": "pending",
                "description": "pending",
                "activity": r['Experiment Group'],
                "alias": [],
                "minimum-number-of-years": r['Size (years, minimum)'],
                "model-realms": [
                    {
                    "id": "pending",
                    "is-required": True
                    }
                ],
                "parent-experiment": [
                    "pending"
                ],
                "start-date": "pending",
                "tier": "pending",
                "@context": "@context",
                "type": [
                    "wcrp:experiment",
                    "cmip7"
                ]
                }
        newfile  = f'experiment/{experiment}_PENDING.json'
        with open(newfile,'w') as f:
            json.dump(template,f,indent=4)  
            
        continue
    
    content = json.load(open(file))
    
    content['activity'] = r['Experiment Group']
    content['minimum-number-of-years'] = r['Size (years, minimum)']
    
    with open(file,'w') as f:
        json.dump(content,f,indent=4)
```
