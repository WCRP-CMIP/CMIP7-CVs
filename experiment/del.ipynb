```python
import re,json,glob,cmipld
```

```python
files = glob.glob('*.json')
```

```python
for f in files:
    with open(f) as fp:
        jd = json.load(fp)
    
    doi = cmipld.utils.texttools.get_doi(str(jd))
    
    jd['references'] = doi
    
    with open(f, 'w') as fp:
        json.dump(jd, fp, indent=2)
```
